```python
def read_data(path,type='csv'):
    if type == "csv":
        return spark.read.format(type).option('inferSchema',True).option('header',True).load(path)
    elif type == "delta":
        return spark.read.load(path)
    elif type=="parquet":
        return spark.read.parquet(path)
    

def write_data(df,path, type='parquet'):
    if type=="parquet":
        print(f"writing data to the path {path}")
        df.write.format('parquet').mode('overwrite').save(path)
        print(f'saving successful!')
    elif type=='delta':
        print(f"writing data to the path {path}")
        df.write.format('delta').mode('overwrite').save(path)
        print(f'saving successful!')


```
